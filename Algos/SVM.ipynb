```python
import numpy as np
class SMV_Classifier ():
    def __init__(self,learning_rate , no_of_terms , lamda_parameter):
        self.learning_rate = learning_rate
        self.no_of_terms = no_of_terms
        self.lamda_parameter = lamda_parameter
    def fit(self,X,Y):
        self.m,self.n = X.shape
        self.w=np.zeros(self.n)
        self.b=0
        self.X=X
        self.Y=Y
        for i in range(self.no_of_terms):
            self.update_weights()
    def update_weights(self):
        y_label = np.where(self.Y<=0,-1,1)
        for index, x_i in enumerate(self.X):
            condition = y_label[index]*(np.dot(x_i,self.w)-self.b)>=1
            if (condition == True):
                dw = 2*self.lamda_parameter*self.w
                db = 0
            else:
                dw = 2*self.lamda_parameter*self.w - np.dot(x_i,y_label[index])
                db = y_label[index]
        self.w = self.w-self.learning_rate*dw
        self.b = self.b-self.learning_rate*db
    def predict(self,X):
        output = np.dot(X,self.w)-self.b
        predicted_labels = np.sign(output)
        y_hat = np.where(predicted_labels<=-1,0,1)
        return y_hat
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(features,target,test_size = 0.2 , random_state = 2)
```

```python
print(features.shape,X_train.shape,X_test.shape)
```

```python
model = SMV_Classifier(0.001,1000,0.01)
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(Y_test , X_test_prediction)
```

```python
print(test_accuracy)
```
